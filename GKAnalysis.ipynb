```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
```

```python
basic_url = "https://fbref.com/en/comps/Big5/keepers/players/Big-5-European-Leagues-Stats"
advanced_url = "https://fbref.com/en/comps/Big5/keepersadv/players/Big-5-European-Leagues-Stats"
```

```python
response1 = requests.get(basic_url).text.replace('<!--', '').replace('-->', '')
response2 = requests.get(advanced_url).text.replace('<!--', '').replace('-->', '')

dfb = pd.read_html(response1, header=1)[1]
dfa = pd.read_html(response2, header=1)[1]

rows_to_keep = []
for i in range(dfb.shape[0]):
    if str(dfb.iloc[i, 0]) != "Rk":
        rows_to_keep.append(i)

dfb = dfb.iloc[rows_to_keep, :].reset_index(drop=True)
dfa = dfa.iloc[rows_to_keep, :].reset_index(drop=True)
```

```python
dfb
```

```python
dfa
```

```python
dfa['PSxG/SoT'] = dfa['PSxG/SoT'].astype(float)
dfb['exs%'] = 0
dfb['exs%'] = dfb['exs%'].astype(float)

for i in range(dfb.shape[0]):
    ex = dfa.iloc[i, dfa.columns.get_loc('PSxG/SoT')]
    dfb.iloc[i, dfb.columns.get_loc('exs%')] = 100 * (1 - ex)
```

```python
dfb['GA-xGA'] = np.nan

dfb['Save%'] = dfb['Save%'].astype(float)

for i in range(dfb.shape[0]):
    a = dfb.iloc[i, dfb.columns.get_loc('Save%')]
    xa = dfb.iloc[i, dfb.columns.get_loc('exs%')]
    dfb.iloc[i, dfb.columns.get_loc('GA-xGA')] = a - xa
    
    if dfb.iloc[i, dfb.columns.get_loc('Comp')] == "de Bundesliga":
        dfb.iloc[i, dfb.columns.get_loc('GA-xGA')] += 2.7
        
    if dfb.iloc[i, dfb.columns.get_loc('Comp')] == "eng Premier League":
        dfb.iloc[i, dfb.columns.get_loc('GA-xGA')] += 2.7
        
    if dfb.iloc[i, dfb.columns.get_loc('Comp')] == "fr Ligue 1":
        dfb.iloc[i, dfb.columns.get_loc('GA-xGA')] += 1.3
        
    if dfb.iloc[i, dfb.columns.get_loc('Comp')] == "es La Liga":
        dfb.iloc[i, dfb.columns.get_loc('GA-xGA')] += 0.6
    
    if dfb.iloc[i, dfb.columns.get_loc('Comp')] == "it Serie A":
        dfb.iloc[i, dfb.columns.get_loc('GA-xGA')] += 2.2
        
dfb['GA-xGA'] = dfb['GA-xGA'].round(4)
```

```python
dfb
```

```python
dfb = dfb.sort_values('GA-xGA', ascending=False)
```

```python
dfb['MP'] = dfb['MP'].astype(float)
rows_to_keep = []
for i in range(dfb.shape[0]):
    if dfb.iloc[i, dfb.columns.get_loc('MP')] >= 10:
        rows_to_keep.append(i)

dfb = dfb.iloc[rows_to_keep, :].reset_index(drop=True)
```
